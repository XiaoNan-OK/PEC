```python
import qutip
from qutip import Qobj,basis
from qutip import sigmax , sigmaz , sigmay
from qutip import tensor
from numpy.random import normal
import numpy as np
```

```python
sigx = sigmax()
sigz = sigmaz()
sigy = sigmay()
iid = qutip.qeye(2)
```

```python
CXH = tensor((iid - sigz) , (iid - sigx))
CX = (-np.pi * 0.25 *  1j *CXH).expm()
```
