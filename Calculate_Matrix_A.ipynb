```python
import math
import numpy as np 
from numpy.linalg import inv
import openpyxl
import cmath
from array import array
```

```python
#Define four initial Density matrix
zero = np.array([[1, 0], [0, 0]])
one = np.array([[0, 0], [0, 1]])
plus = np.array([[0.5, 0.5], [0.5, 0.5]])
right = np.array([[0.5, -0.5j], [0.5j, 0.5]])
init = {'0':zero, '1':one, '+':plus, 'R':right}

#Define observable
I = np.array([[1, 0], [0, 1]])
X = np.array([[0, 1], [1, 0]])
Y = np.array([[0, -1j], [1j, 0]])
Z = np.array([[1, 0], [0, -1]])
obs = {'I':I, 'X':X, 'Y':Y, 'Z':Z}

#Calculate two qubits initial density matrix
Initial = {}
for findex, fmatrix in init.items():
    for sindex, smatrix in init.items():
        Initial.update({f'{findex}{sindex}':np.kron(fmatrix, smatrix)})

#Calculate two qubits observable
Observable = {}
for findex, fmatrix in obs.items():
    for sindex, smatrix in obs.items():
        Observable.update({f'{findex}{sindex}':np.kron(fmatrix, smatrix)})
```

```python
#Calculate A matrix
A = []
for findex, fmatrix in Initial.items():
    each = []
    for sindex, smatrix in Observable.items():
        each.append(np.trace(np.matmul(fmatrix, smatrix)))
    A.append(each)

A = np.transpose(A)
A
```

```python
Areal = A.real.astype(int)
```

```python
import numpy as np
import csv

# 將矩陣存成CSV檔案
filename = 'matrix_A.csv'
np.savetxt(filename, Areal, delimiter=",", fmt='%d')

print(f"矩陣已成功存成CSV檔案：{filename}")
```
